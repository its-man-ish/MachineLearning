now I have values like (x, x1-x2, x.0) and some other wiered numbers so I am converting x into float and taking avergae of x1-x2 and removing wiered values

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

 this function will return all the rows in totsal sqft colnm in which we dont havve valid float values

```python
df3[~df3['total_sqft'].apply(is_float)] 
```

This is typical datascince problem where data is not structured and messy, it has out liers errors, so majority of the time data scientist do is data cleaning

```python
def covert_sqft_to_num(x):
    tokens=x.split('-')
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
covert_sqft_to_num('211') # this convert string into float
```

```python
covert_sqft_to_num('2100-2850') # avg of the two numbers
```

```python
# now wil apply this function on dataset
```

```python
df4=df3.copy()
```

```python
df4['total_sqft']=df4['total_sqft'].apply(covert_sqft_to_num)
                                    
```
