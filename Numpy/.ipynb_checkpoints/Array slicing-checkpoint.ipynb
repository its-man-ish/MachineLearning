# Concatenition

```python
mtx1=np.arange(1,101).reshape(10,10)
mtx2=np.arange(30,130).reshape(10,10)
```

```python
np.concatenate((mtx1,mtx2))
```

```python
np.concatenate((mtx1,mtx2),axis=1)
```
